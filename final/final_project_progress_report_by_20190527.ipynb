## 主題
參加[新聞立場檢索技術獎金賽](https://aidea-web.tw/topic/b6abbf14-2d60-456c-8cbe-34fdfcd58967)
## 介紹
參賽者需能夠透過輸入搜尋關鍵字，找出新聞資料庫中最相關的新聞，同時必須要能夠分辨新聞的立場，使得關鍵字的立場能夠與新聞立場高度相關。
主辦單位並提供數個搜尋關鍵詞，以及與該搜尋關鍵詞相關的新聞及其「相關程度排名」；同時也有「相關程度分數」。
資料來源為各大新聞網站（包含四大報如蘋果日報、自由時報、聯合報、中國時報以及其他電子媒體如TVBS等新聞）等，參賽者需自行爬下新聞內容。
## 實作方法

* 資料預處理
    * 將文字分詞後，對每一則新聞建立TF-IDF矩陣，瞭解每則新聞中各個詞語的詞頻與詞重要性。
    * 使用情緒分析（sentiment analysis）方式建立文件情緒的「情緒分數」，也就相當於可以貼近新聞的立場。
    * 將每個詞依據word2vec方式轉換為向量。word2vec可以將詞以向量（word-embedding）的方式表現，也因此詞與詞在多維空間中的距離差距可以衡量。此部分需要蒐羅大量新聞文章建立語料庫後才能訓練得出一個適合的詞向量。
* 學習方式
    * 輸入（自變項）：先以代碼、數值、向量等方式表示搜尋關鍵字與新聞文章詞語、詞頻、詞重要性分數、情感分數。接著計算搜尋關鍵字與新聞文章詞語之間的「距離」。 
    * 輸出（應變項）：排名（轉換為分數）、相關分數。
    * 函式： f(搜尋關鍵字, 新聞文章) = 距離／相似性／相關排名／相關分數。
    * 模型選擇以RNN(LSTM)及CNN皆可，須經試用並比較。
