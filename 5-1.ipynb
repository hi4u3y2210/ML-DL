```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import train_test_split

file_path = "5-1.csv"
data = pd.read_csv(file_path, encoding = "utf8", delimiter = "\t")
data = data[:216]
```

```python
price = np.array(data.price)
WTI = np.array(data.WTI)
SP500 = np.array(data.SP500)
gold_london = np.array(data.gold_london)
unemployment_rate = np.array(data.unemployment_rate)
fx = np.array(data.fx)
y = price.reshape(len(price), 1)
```

```python
regr = LinearRegression()
```

台股跟原油(WTI)

```python
X_WTI = WTI.reshape(len(WTI), 1)
regr.fit(X_WTI, price)
Y = regr.predict(X_WTI)
plt.scatter(WTI, price)
plt.plot(WTI, Y, "r")
```

```python
regr.score(X_WTI,price)
```

```python
X_SP500 = SP500.reshape(len(SP500), 1)
regr.fit(X_SP500, price)
Y = regr.predict(X_SP500)
plt.scatter(SP500, price)
plt.plot(SP500, Y, "r")
```

```python
regr.score(X_SP500,price)
```
